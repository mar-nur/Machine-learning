### Визуализация распределения переменных

#### Гистограмма для числовых переменных

```python
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
n_cols = len(num_cols)
n_rows = (n_cols + 2) // 3 
    
fig, axes = plt.subplots(n_rows, 3, figsize=(18, 4 * n_rows))
axes = axes.flatten()
    
for i, col in enumerate(num_cols):
    sns.histplot(df[col], kde=True, ax=axes[i])
    axes[i].set_title(f'Распределение {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Частота')
    
plt.tight_layout()
plt.show()
```

#### Распределения, близкие к нормальным:

- Sleep Duration (продолжительность сна) - приближается к нормальному распределению
- Quality of Sleep (качество сна) - умеренно близко к нормальному
- Heart Rate (пульс) - относительно нормальное распределение

#### Распределения, отличающиеся от нормального:

- Person ID - равномерное распределение (что ожидаемо для идентификаторов)
- Age (возраст) - бимодальное распределение с концентрациями вокруг 30 и 50 лет
- Physical Activity Level - правостороннее асимметричное распределение
- Stress Level - левостороннее асимметричное распределение
- Daily Steps - распределение с несколькими пиками

#### Концентрация большинства значений:
- Age: два основных кластера - около 30 лет и около 50 лет
- Sleep Duration: концентрация между 6.5-7.5 часами
- Quality of Sleep: пик вокруг значений 7-8 баллов
- Physical Activity Level: большинство значений 40-60 минут/день
- Stress Level: концентрация в диапазоне 4-7 баллов
- Heart Rate: нормальное распределение вокруг 70-75 уд/мин
- Daily Steps: несколько пиков, наиболее выраженный около 7000 шагов

#### Предсказания
- Sleep Disorder vs Sleep Duration/Quality: как нарушения сна влияют на продолжительность и качество сна
- Gender vs все числовые показатели: гендерные различия в параметрах сна и образа жизни
- Occupation vs Stress Level/Sleep Quality: как профессия влияет на уровень стресса и качество сна

### Столбчатая диаграмма для категориальной переменной Class

```python
cat_cols = df.select_dtypes(include=['category', 'object']).columns
n_cols = len(cat_cols)
n_rows = (n_cols + 2) // 3  # Округление вверх до ближайшего целого
    
fig, axes = plt.subplots(n_rows, 3, figsize=(18, 4 * n_rows))
axes = axes.flatten()
    
for i, col in enumerate(cat_cols):
    sns.countplot(data=df, x=col, ax=axes[i])
    axes[i].set_title(f'Распределение {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Частота')
        
    axes[i].tick_params(axis='x', rotation=45)
    
plt.tight_layout()
plt.show()
```

#### Occupation (Профессия)

- Наибольшая концентрация в категории "Nurse" (Медсестра)
- Второе место - "Engineer" (Инженер)
- Остальные профессии представлены значительно реже

#### BMI Category (Категория ИМТ)

- Наиболее распространенная категория - "Normal" (Нормальный вес)
- "Overweight" (Избыточный вес) занимает второе место
- Категории "Obese" (Ожирение) и "Normal Weight" представлены примерно одинаково

#### Sleep Disorder (Расстройство сна)

- Большинство наблюдений не имеют расстройств сна ("None")
- Среди имеющих расстройства преобладает "Sleep Apnea" (Апноэ сна)
- "Insomnia" (Бессонница) встречается реже всего

#### Blood Pressure (Артериальное давление)

- Наиболее частые значения группируются вокруг нормальных показателей
- Есть несколько выраженных пиков в распределении

#### Gender (Пол)

- Примерно равномерное распределение между мужчинами и женщинами

#### Stress Level (Уровень стресса)

- Наименьшее количество наблюдений на экстремальных значениях (очень низкий и очень высокий стресс)

### Боксплот для числовой переменной

```python
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
n_cols = len(num_cols)
n_rows = (n_cols + 2) // 3 
    
fig, axes = plt.subplots(n_rows, 3, figsize=(18, 4 * n_rows))
axes = axes.flatten()

for ax, col in zip(axes, num_cols):
   sns.boxplot(df[col], ax=ax)
   ax.set_title(f'Распределение {col}')
   ax.set_xlabel(col)
   ax.set_ylabel('Частота')

plt.tight_layout()
plt.show()
```

Выбросы присутствуют на распределении боксплота пульса (Heart Rate), может быть связано с сердечно-сосудистыми зааболеваниями людей, так как медианное значения пульса 70.

Анализируя медианы по категориям, можно выявить:

- Различия в продолжительности сна между людьми с разными нарушениями сна
- Разный уровень стресса в зависимости от профессии или пола
- Отличия в физической активности между различными BMI категориями

### Корреляционная матрица

```python
# Корреляционная матрица
numeric_df = df.select_dtypes(include=['int64', 'float64'])
correlation_matrix = numeric_df.corr()

# Визуализация тепловой карты
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Корреляционная матрица')
plt.show()
```

- Sleep Duration и Quality of Sleep (количество часов сна и оценка сна)
